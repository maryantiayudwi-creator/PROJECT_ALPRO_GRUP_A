```python


def analisis_teks():

    teks = input("Masukkan teks: ")


    if not teks.strip():
        jumlah_kata = 0
        jumlah_kalimat = 0
    else:

        jumlah_kata = len(teks.split())


        jumlah_kalimat = len(teks.split(".")) - 1

    print(f"\nTeks tersebut memuat {jumlah_kalimat} kalimat dan {jumlah_kata} kata.")

if __name__ == "__main__":
    analisis_teks()
```
