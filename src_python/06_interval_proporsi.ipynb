```python
import math

def hitung_interval_konfidensi(p_hat, alpha, n):
    if p_hat < 0 or p_hat > 1:
        return "Error: Proporsi harus berada pada rentang 0 sampai 1."

    if alpha == 0.10:
        z = 1.645
    elif alpha == 0.05:
        z = 1.96
    else:
        return "Error: Alpha harus 0.10 atau 0.05."

    margin_error = z * math.sqrt((p_hat * (1 - p_hat)) / n)

    batas_bawah = p_hat - margin_error
    batas_atas = p_hat + margin_error

    return batas_bawah, batas_atas

# Input pengguna
p_hat = float(input("Masukkan proporsi sampel (p̂): "))
alpha = float(input("Masukkan alpha (0.10 atau 0.05): "))
n = int(input("Masukkan ukuran sampel (n): "))

hasil = hitung_interval_konfidensi(p_hat, alpha, n)

if isinstance(hasil, str):
    print(hasil)
else:
    batas_bawah, batas_atas = hasil

    print("\nInterval Konfidensi:")
    print(f"({batas_bawah:.4f}, {batas_atas:.4f})")
```
