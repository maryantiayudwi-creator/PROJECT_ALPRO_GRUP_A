```python

import math

def hitung_akar_kuadrat():
    print("--- Program Perhitungan Akar Persamaan Kuadrat ---")
    a = float(input("Masukkan koefisien a: "))
    b = float(input("Masukkan koefisien b: "))
    c = float(input("Masukkan koefisien c: "))

    if a == 0:
        print("Bukan persamaan kuadrat karena nilai 'a' tidak boleh 0!")
        return

    D = (b ** 2) - (4 * a * c)
    print(f"Nilai Diskriminan (D) = {D}")

    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2 * a)
        x2 = (-b - math.sqrt(D)) / (2 * a)
        print(f"Persamaan memiliki dua akar real berbeda:")
        print(f"x1 = {x1:.3f}")
        print(f"x2 = {x2:.3f}")
    elif D == 0:
        x = -b / (2 * a)
        print(f"Persamaan memiliki satu akar real kembar:")
        print(f"x = {x:.3f}")
    else:
        print("Persamaan tersebut hanya memiliki akar-akar imajiner.")

if __name__ == "__main__":
    hitung_akar_kuadrat()
```
