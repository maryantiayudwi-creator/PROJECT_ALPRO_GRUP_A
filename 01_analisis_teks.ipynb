```python
def analisis_tiga_input():
    print("--- PROGRAM ANALISIS 3 KASUS TEKS --- \n")


    for i in range(1, 4):
        teks = input(f"Masukkan teks Kasus {i}: ")

        if not teks.strip():
            jumlah_kata1 = 0
            jumlah_kalimat1 = 0
        else:
            jumlah_kata1 = len(teks.split())
            jumlah_kalimat1 = len(teks.split(".")) - 1

        print(f"-> Hasil Kasus {i}: Memuat {jumlah_kalimat1} kalimat dan {jumlah_kata1} kata.\n")


if __name__ == "__main__":
    analisis_tiga_input()
```
