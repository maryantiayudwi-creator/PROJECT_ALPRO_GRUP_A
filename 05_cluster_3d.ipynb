```python
import math

def hitung_jarak(titik1, titik2):
    return math.sqrt(
        (titik1[0] - titik2[0])**2 +
        (titik1[1] - titik2[1])**2 +
        (titik1[2] - titik2[2])**2
    )

def tentukan_cluster(titik_u):
    cluster_a = (2, 1, 3)
    cluster_b = (1, -4, 6)
    cluster_c = (-2, 3, -2)

    jarak_a = hitung_jarak(titik_u, cluster_a)
    jarak_b = hitung_jarak(titik_u, cluster_b)
    jarak_c = hitung_jarak(titik_u, cluster_c)

    if jarak_a <= jarak_b and jarak_a <= jarak_c:
        return "Cluster A"
    elif jarak_b <= jarak_a and jarak_b <= jarak_c:
        return "Cluster B"
    else:
        return "Cluster C"

# Input koordinat titik U
print("Masukkan koordinat titik U(x1, x2, x3)")
x1 = float(input("x1 = "))
x2 = float(input("x2 = "))
x3 = float(input("x3 = "))

titik_u = (x1, x2, x3)

hasil = tentukan_cluster(titik_u)

print("\nTitik U =", titik_u)
print("Titik tersebut termasuk", hasil)
```
