```text


START
    PRINT "--- PROGRAM ANALISIS 3 KASUS TEKS ---"
    

    FOR i ← 1 TO 3 DO
        PRINT "Masukkan teks Kasus ", i, ": "
        READ teks

        IF STRIP(teks) == "" THEN
            jumlah_kata ← 0
            jumlah_kalimat ← 0
        ELSE
            jumlah_kata ← LENGTH(SPLIT(teks, " "))
            jumlah_kalimat ← LENGTH(SPLIT(teks, ".")) - 1
        ENDIF

        PRINT "-> Hasil Kasus ", i, ": Memuat ", jumlah_kalimat, " kalimat dan ", jumlah_kata, " kata."
        PRINT "--------------------------------------------------"
    ENDFOR
END
