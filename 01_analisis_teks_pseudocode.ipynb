```text
NOMOR 1

START
    PRINT "Masukkan teks: "
    INPUT teks
    
    IF NOT teks.strip() THEN
        jumlah_kata = 0
        jumlah_kalimat = 0
    ELSE
        jumlah_kata = len(teks.split())
        jumlah_kalimat = len(teks.split(".")) - 1
    ENDIF
    
    PRINT "Teks tersebut memuat ", jumlah_kalimat, " kalimat dan ", jumlah_kata, " kata."
END
