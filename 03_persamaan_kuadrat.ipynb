```python
import math

def hitung_akar_kuadrat():

    a = float(input("Masukkan nilai koefisien a: "))
    b = float(input("Masukkan nilai koefisien b: "))
    c = float(input("Masukkan nilai koefisien c: "))

    if a == 0:
        print("Nilai 'a' tidak boleh 0 karena ini bukan persamaan kuadrat.")
        return

    D = b**2 - 4*a*c

    print(f"\nNilai Diskriminan (D) = {D}")

    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        print("Persamaan memiliki DUA AKAR REAL yang berbeda:")
        print(f"x1 = {x1:.3f}")
        print(f"x2 = {x2:.3f}")

    elif D == 0:
        x = -b / (2*a)
        print("Persamaan memiliki SATU AKAR REAL (kembar):")
        print(f"x = {x:.3f}")

    else:
        print("Persamaan tersebut HANYA memiliki akar-akar imajiner.")

if __name__ == "__main__":
    hitung_akar_kuadrat()
```
